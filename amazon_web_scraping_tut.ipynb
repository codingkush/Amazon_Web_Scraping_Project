```python
# Load environment variables from the .env file
load_dotenv()

# Access the variables
USER_AGENT = os.getenv("USER_AGENT")
ACCEPT_LANGUAGE = os.getenv("ACCEPT_LANGUAGE")
# print(USER_AGENT)
# print(ACCEPT_LANGUAGE)

#use user_agent and accept_langauage to hide your user agent publicly
```

### Basics for web scraping

```python
URL = "https://www.amazon.in/s?k=playstation+5&crid=3O8SOC5QG6OIU&sprefix=playstation+5%2Caps%2C237&ref=nb_sb_noss_2"
```

```python
#Headers for requests 
#EX: 
# HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36',
#     'Accept-Language': 'en-US,en;q=0.5'})

HEADERS = ({'User-Agent':USER_AGENT,'Accept-Language':ACCEPT_LANGUAGE})
```

```python
#HTTP requests 
webpage = requests.get(URL,headers=HEADERS)
```

```python
webpage
```

```python
webpage.content
```

```python
type(webpage.content)
```

```python
#Soup object containing all data 
soup = BeautifulSoup(webpage.content,"html.parser")
soup
```

```python
#Fetch links as List of Tag Objects 

#find all find all a tags (achor tag) and extracted
links = soup.find_all("a",attrs={'class':'a-link-normal s-line-clamp-2 s-link-style a-text-normal'})
```

```python
links
```

```python
# links[0]
link= links[0].get('href')
```

```python
product_list = "https://amazon.in" + link
product_list
```

```python
#HTTP requests
new_webpage = requests.get(product_list,headers=HEADERS)
```

```python
new_webpage
```

```python
#Soup object containing all data 
new_soup = BeautifulSoup(new_webpage.content,"html.parser")
new_soup
```

```python
new_soup.find("span",attrs={'id':'productTitle'}).text.strip()
```
